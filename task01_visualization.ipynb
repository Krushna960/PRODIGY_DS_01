```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("world_population.csv")
```

```python
top10 = df.sort_values(by='2022 Population', ascending=False).head(10)
```

```python
POP_COLUMN = '2022 Population'  # example

top10 = df.sort_values(by=POP_COLUMN, ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10['Country'], top10[POP_COLUMN] / 1e6)
plt.title("Top 10 Most Populated Countries")
plt.xlabel("Country")
plt.ylabel("Population (Millions)")
plt.xticks(rotation=45)
plt.show()
```

```python
top5 = df.sort_values(by='2022 Population', ascending=False).head(5)
```

```python
plt.figure(figsize=(7,7))
plt.pie(
    top5['2022 Population'],
    labels=top5['Country'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Population Share of Top 5 Countries")
plt.show()
```

```python
india = df[df['Country'] == 'India']
india
```

```python
years = ['1970 Population', '1990 Population', '2000 Population', '2015 Population', '2022 Population']
india_pop = india[years].T
india_pop.columns = ['Population']

plt.figure(figsize=(8,5))
plt.bar(india_pop.index, india_pop['Population'] / 1e6)
plt.title("Population Growth of India")
plt.xlabel("Year")
plt.ylabel("Population (Millions)")
plt.show()
```
